```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#the cancer dataset
bcancer_data = load_breast_cancer()
```

```python
bcancer_df = pd.DataFrame(bcancer_data.data)
bcancer_df.head()
```

```python
bcancer_df.shape
```

```python
bcancer_df.columns = bcancer_data.feature_names
bcancer_df.head()
```

```python
bcancer_df.shape
```

```python
bcancer_df['CLASS'] = bcancer_data.target
bcancer_df.head()
```

```python
import seaborn as sns
from sklearn.model_selection import cross_val_score
```

```python
sns.countplot(x=bcancer_df['CLASS'])
plt.show()
```

```python
correlation =bcancer_df.corr()
plt.figure(figsize=(14,14))
sns.heatmap(correlation,cbar=True,square=True,cmap='coolwarm')
plt.show()
```

```python
bcancer_df.drop('CLASS',axis=1)
```

```python
#data for the model
x = bcancer_df.drop('CLASS',axis=1)
y = bcancer_df['CLASS']
```

```python
print([x.shape,y.shape])
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
print([X_train.shape,y_train.shape])
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtmodel = DecisionTreeClassifier()
```

```python
dtmodel
```

```python
dtmodel.fit(X_train,y_train)
```

```python
y_pred = dtmodel.predict(X_test)
```

```python
y_df = pd.DataFrame(y_pred,y_test)
y_df.head()
```

```python
y_df.tail(10)
```

```python
dtmodel.score(X_test,y_test)
```

```python
cross_val_score(dtmodel,X_test,y_test)
```
